```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cross_decomposition import CCA
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
X = pd.read_csv('df1.csv')
Y = pd.read_csv('df2.csv')
```

```python
#Standardize X
scaler1 = StandardScaler()
X[[X.columns[i] for i in range(3)]] = scaler1.fit_transform(X[[X.columns[i] for i in range(3)]])
```

```python
# Standardize Y
scaler2 = StandardScaler()
Y[[Y.columns[i] for i in range(3)]] = StandardScaler().fit_transform(Y[[Y.columns[i] for i in range(3)]])
```

```python
X[:5]
```

```python
Y[:5]
```

```python
for i in range(1, 8):
    cca = CCA(n_components=i)
    cca.fit(X, Y)
    X_c, Y_c = cca.transform(X, Y)
    canonical_corr = [np.corrcoef(X_c[:, i], Y_c[:, i])[0, 1] for i in range(i)]

    # Output canonical correlations
    print(f"n_components == {i}: CCA score is {cca.score(X, Y)}")
    print("Canonical Correlations:", canonical_corr)
    print()
    
```
